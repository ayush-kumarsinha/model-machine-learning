## Decision Tree Classifier Implementation With Post Preprunning

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

We are using here load_iris dataset

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

WE CAN IMPORT ALSE SEABORN DATASET

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
dataset.target
```

Independent and dependent

```python
X=df.iloc[:,:-1]
y=dataset.target
```

```python
X,y
```

## Train Test Splite


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

Post Pruning

```python
treeclassifier=DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True)
plt.show()
```

```python
## Post Pruning with max_depth =2
treeclassifier=DecisionTreeClassifier(criterion='entropy',max_depth=2)
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## prediction
y_pred=treeclassifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```
