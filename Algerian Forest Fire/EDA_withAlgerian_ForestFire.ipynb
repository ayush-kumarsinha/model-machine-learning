# Algerian Forest Fires Dataset

## Data Set Information
The dataset includes 244 instances that regroup data of two regions of Algeria:
Bejaia region (northeast of Algeria) and Sidi Bel-abbes region (northwest of Algeria).
122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attributes and 1 output attribute (class).
The 244 instances have been classified into fire (138) and not fire (106) classes.

## Attribute Information
1. Date (DD/MM/YYYY)
2. Temp (temperature in Celsius)
3. RH (Relative Humidity)
4. Ws (Wind speed km/h)
5. Rain (mm)
6. FFMC
7. DMC
8. DC
9. ISI
10. BUI
11. FWI
12. Classes (Fire / Not Fire)
